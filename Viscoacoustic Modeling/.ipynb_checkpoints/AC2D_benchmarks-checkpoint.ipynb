```python
S1 = Python_CPU_Numba[-1]/Python_CPU_Numba[-1]
S2 = Python_CPU_Numba[-1]/Python_CPU_Cython[-1]
S3 = Python_CPU_Numba[-1]/C_CPU[-1]
S4 = Python_CPU_Numba[-1]/Julia_CPU[-1]

S5 = Python_CPU_Numba[-1]/Python_Multithread_Numba[-1]
S6 = Python_CPU_Numba[-1]/Python_Multithread_Cython[-1]
S7 = Python_CPU_Numba[-1]/Julia_Multithread[-1]
S8 = Python_CPU_Numba[-1]/Julia_Multithread[-1]

S9 = Python_CPU_Numba[-1]/Python_Multithread_Numba_CuPy[-1]
S10 = Python_CPU_Numba[-1]/C_GPU[-1]
S11 = Python_CPU_Numba[-1]/Julia_GPU[-1]

# Labels for the bars
labels = ['Numba CPU', 'Cython CPU', 'C CPU', 'Julia CPU', 
          'Numba Multithread', 'Cython Multithread', 'C Multithread', 
          'Julia Multithread', 'Numba CuPy GPU', 'C GPU', 'Julia GPU']

# Runtime values in a list
runtime = [Python_CPU_Numba[-1], Python_CPU_Cython[-1], C_CPU[-1], Julia_CPU[-1], 
           Python_Multithread_Numba[-1], Python_Multithread_Cython[-1], Julia_Multithread[-1], 
           Julia_Multithread[-1], Python_Multithread_Numba_CuPy[-1], C_GPU[-1], Julia_GPU[-1]]

# Speedup values in a list
speedup = [S1, S2, S3, S4, S5, S6, S7, S8, S9, S10, S11]

# Define a list of colors
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'purple', 'brown', 'pink', 'orange']

import matplotlib.cm as cm

# Normalize the speedup values to the range 0-1
#normalized_speedup = [(spd - min(speedup)) / (max(speedup) - min(speedup)) for spd in speedup]
normalized_speedup = [0.7 + 0.7 * (spd - min(speedup)) / (max(speedup) - min(speedup)) for spd in speedup]

# Create a colormap
cmap = cm.get_cmap('Greens')

plt.figure(figsize=(15, 9))

# Create a bar for each label, runtime, and speedup
for i, (label, run, spd, norm_spd) in enumerate(zip(labels, runtime, speedup, normalized_speedup)):
    # Create a bar with a specific color and label
    color = cmap(norm_spd)
    bar = plt.bar(i, run, color=color, label=label, zorder=2)
    
    # Display the speedup above the bar
    yval = bar[0].get_height()
    plt.text(bar[0].get_x() + bar[0].get_width()/2, yval + 0.05, f'{round(spd, 2)}X', ha='center', va='bottom', fontsize=20)

# Add labels and title
plt.xlabel(' ')
plt.ylabel('Runtime (s)', fontsize = 15)
plt.title('Speedup for the 6000x6000 model', fontsize = 20)

# Remove x-axis labels
#plt.xticks([])
plt.xticks(range(len(labels)), labels, rotation=45, fontsize = 14)

# Add gridlines
plt.grid()

# Add legend
#plt.legend()


# Display the plot
plt.tight_layout()
plt.savefig("Performance6.svg", format="svg", bbox_inches = 'tight')
plt.show()
```
